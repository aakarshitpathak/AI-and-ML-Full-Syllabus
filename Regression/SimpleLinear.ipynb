```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

# Linear Regression

* relationship between the dependent and independent variable
* involves only one independent variable
# Note:-
* models are only trained in 2D arrays

```python
x= df[['YearsExperience']] #double brackets to make it a 2D array 
# x is using for features
y= df['Salary'] # single brackets to make it a 1D array
# y is using for labels
```

```python
X_train, X_test, y_train , y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

# Model Training

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
# Making a single prediction (for example the salary of an employee with 12 years of experience)
print(regressor.predict([[12]]))
```

```python
plt.scatter (X_train,y_train, color = 'red' )
plt.plot(X_train , regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train , regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
from joblib import dump, load
with open('models/simple_linear_regression_model.joblib','wb') as f:
    dump(regressor, f) #dump the model to the file
    print("Model Dumped Successfully")
```
