```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y, color='blue', label="Data Points")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
